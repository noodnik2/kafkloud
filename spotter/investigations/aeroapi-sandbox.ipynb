```python
from dotenv import dotenv_values

API_KEY = dotenv_values(".env.local")["AEROAPI_API_KEY"]

AUTH_HEADER = {
    "x-apikey": API_KEY,
}

QUERY = {"query": '-latlong "44.953469 -111.045360 40.962321 -104.046577"'} // example response in: latlong.jsno

import requests
URLBASE_AEROAPI = "https://aeroapi.flightaware.com/aeroapi"
print(requests.get(URLBASE_AEROAPI + "/flights/search", headers=AUTH_HEADER, params=QUERY).json())
```

```python
from dotenv import dotenv_values
import requests
AEROAPI = requests.Session()
AEROAPI.headers.update({"x-apikey": dotenv_values(".env.local")["AEROAPI_API_KEY"]})
print(
    AEROAPI.get(
        "https://aeroapi.flightaware.com/aeroapi/airports/KSMO/flights",
        params={
            "type": "General_Aviation",
            "start": "2023-05-05T21:44:30Z"
        }
    ).json()
)
```
